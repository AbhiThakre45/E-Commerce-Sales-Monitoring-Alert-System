```python
### Category wise Revenue 

pd.read_sql("Select Category,sum(Revenue) as Total_Revenue from retail_cleaned_data Group by Category order by Total_Revenue Desc", engine)
```

```python
### Category wise profit 

pd.read_sql("Select Category, sum(Profit) as Total_Profit from retail_cleaned_data Group by Category Order by Total_Profit Desc", engine)
```

```python
### Region wise Revenue

pd.read_sql("Select Region, sum(Revenue) as Total_Revenue from retail_cleaned_data Group by Region", engine)
```

```python
### Top 5 Product 

pd.read_sql("Select Product_name,sum(Revenue) as Revenue from retail_cleaned_data Group by Product_Name Order by Revenue Desc Limit 5", engine)
```

```python
### Demand Gap 

pd.read_sql("Select Product_Name , Demand_Forecast, Units_Sold,(Demand_Forecast - Units_Sold) as Demand_Gap from retail_cleaned_data Order by Demand_Gap Desc Limit 5", engine)
```

```python
### Monthly Revenue Trend 

pd.read_sql("Select Month, Sum(Revenue) as Monthly_Revenue from retail_cleaned_data Group by Month Order by Month",engine)
```

```python
### Quartery Revenue Trend 

pd.read_sql("Select Quarter, Sum(Revenue) as Quarterly_Revenue from retail_cleaned_data Group by Quarter Order by Quarter",engine)
```

```python
### Avg Profit Margin 

pd.read_sql("Select AVG(Profit_Margin_pct) as Avg_Profit_Margin from retail_cleaned_data",engine)
```

```python
## Top product by profit 

pd.read_sql("""
SELECT 
    Product_Name,
    total_profit,
    RANK() OVER (ORDER BY total_profit DESC) AS rank_position
FROM (
    SELECT Product_Name, SUM(Profit) AS total_profit
    FROM retail_cleaned_data
    GROUP BY Product_Name
) t;
""", engine)
```

```python
### Running Revenue

pd.read_sql("""
SELECT 
    Month,
    monthly_revenue,
    SUM(monthly_revenue) OVER (ORDER BY Month) AS running_total
FROM (
    SELECT Month, SUM(Revenue) AS monthly_revenue
    FROM retail_cleaned_data
    GROUP BY Month
) t;
""", engine)
```

```python
## High Demand Gap Product 

pd.read_sql("""
SELECT 
    Product_Name,
    (Demand_Forecast - Units_Sold) AS demand_gap
FROM retail_cleaned_data
WHERE (Demand_Forecast - Units_Sold) > 10
ORDER BY demand_gap DESC limit 10
""", engine)
```
